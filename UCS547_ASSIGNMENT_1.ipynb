Ques. 2. Compare the following numpy arrays: arr1 = np.array([[1, 2], [3, 4]])

arr2 = np.array([[1, 2], [3, 4]])

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])

comparison = arr1 == arr2
are_equal = np.array_equal(arr1, arr2)

print("Element-wise comparison:\n", comparison)
print("Are both arrays completely equal?", are_equal)
```

Ques.3 Find the most frequent value and their indice(s) in the following arrays:

i. x = np.array([1,2,3,4,5,1,2,1,1,1])

ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
# Q3(i): Most frequent value and indices in x
x = np.array([1,2,3,4,5,1,2,1,1,1])

values, counts = np.unique(x, return_counts=True)
most_frequent_value = values[np.argmax(counts)]
indices = np.where(x == most_frequent_value)

print("Array x:", x)
print("Most frequent value:", most_frequent_value)
print("Indices:", indices)
```

```python
# Q3(ii): Most frequent value and indices in y
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])

values, counts = np.unique(y, return_counts=True)
most_frequent_value = values[np.argmax(counts)]
indices = np.where(y == most_frequent_value)

print("Array y:", y)
print("Most frequent value:", most_frequent_value)
print("Indices:", indices)
```

Ques.4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),
find

i. Sum of all elements

ii. Sum of all element’s row-wise

iii. Sum of all element’s column-wise

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
total_sum = gfg.sum()

# ii. Row-wise sum
row_wise_sum = gfg.sum(axis=1)

# iii. Column-wise sum
column_wise_sum = gfg.sum(axis=0)

print("Matrix:\n", gfg)
print("Sum of all elements:", total_sum)
print("Row-wise sum:\n", row_wise_sum)
print("Column-wise sum:\n", column_wise_sum)
```
